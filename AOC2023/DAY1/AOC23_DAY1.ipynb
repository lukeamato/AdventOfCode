```python
# PART I
f = open('input1.txt', 'r')

total = 0

for line in f:
  first = None
  last = None
  i = 0
  while not first:
    if line[i].isdigit():
      first = line[i]
    i += 1
  i = len(line) - 1
  while not last:
    if line[i].isdigit():
      last = line[i]
    i -= 1
  final = int(first + last)
  total += final

print(total)
```

```python
# PART II
f = open('input1.txt', 'r')

total = 0

for line in f:
  first = None
  first_index = None
  last = None
  last_index = None
  i = 0
  line_length = len(line)
  while not first and i < line_length:
    if line[i].isdigit():
      first = line[i]
      first_index = i
    i += 1
  i = len(line) - 1
  while not last and i > -1:
    if line[i].isdigit():
      last = line[i]
      last_index = i
    i -= 1
  digits = {'one': '1', 'two': '2', 'three': '3', 'four': '4', 'five': '5', 'six': '6', 'seven': '7', 'eight': '8', 'nine': '9'}
  for dig in digits:
    ind = line.find(dig)
    if ind != -1 and (first_index is None or ind < first_index):
      first = digits[dig]
      first_index = ind
    ind = line.rfind(dig)
    if ind != -1 and (last_index is None or ind > last_index):
      last = digits[dig]
      last_index = ind
  final = int(first + last)
  print(final)
  total += final

print(total)
```
